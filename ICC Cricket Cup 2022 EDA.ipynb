# How Many Time Team Win and Select Bat First vs Bowl First

```python
dt_td=data['Toss_decision'].value_counts()
dt_td
```

```python
plt.figure(figsize=(5,5))
plt.pie(dt_td,labels=dt_td.index,autopct="%1.1f%%",startangle=90)
plt.axis('equal')
plt.title('Toss Decision')
plt.show()
```

## How Many Loss the Toss but win & How Many Team Loss the toss and Loss the Match?

```python
data.head(2)
```

```python
Toss_losser_win=data[data['Toss_winner']!=data['Winner']].shape[0]
Toss_losser_loss=data[data['Toss_winner']==data['Winner']].shape[0]
```

```python
sizes=[Toss_losser_win,Toss_losser_loss]
```

```python
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=['Winer','Loss'], colors=['green','red'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Toss Losser Match Win/Loss Distribution')
```

# How many Time Toss Winner Team Select Bat First and Win and How Many time Toss Winner Choose Field First and Win vice versa

```python
bat_first_data=data[data['Toss_decision']=='Bat']
field_first_data=data[data['Toss_decision']=='Field']
```

```python
bat_first_win=bat_first_data[bat_first_data['Toss_winner']==bat_first_data['Winner']].shape[0]
bat_first_loss=bat_first_data.shape[0]-bat_first_win
```

```python
field_first_win=field_first_data[field_first_data['Toss_winner']==field_first_data['Winner']].shape[0]
field_first_loss=field_first_data.shape[0]-field_first_win
```

```python
field_first_loss
```

```python
xsize=[bat_first_win,bat_first_loss,field_first_win,field_first_loss]
plt.pie(xsize,labels=['Bat Selector Win','Bat Selector Loss','Field Selector Win','Field Selector Loss'],autopct='%1.1f%%',startangle=90)
```
